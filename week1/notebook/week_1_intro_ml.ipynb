### Q-2

String manipulation + Attribute rename

```python
# string manipulation

# rename the species attributes to have I- prefix instead of Iris-

data['species'] = data['species'].str.replace("Iris-", "I-")

# show the updated DataFrame
data.head()
```

### Q-3

Data analysis

```python
# get the number of each species present meaning bins of species

species_counts = data['species'].value_counts()

# display the results
for species, count in species_counts.items():
    print(f" - {species}: {count}")
```

```python
# get the mean, median, ranges, min, max etc. of each numeric attribute


for col in data.select_dtypes(include=['float64', 'int64']).columns:
    print(f"Attribute: {col}")
    print(f" - Mean: {data[col].mean():.2f}")
    print(f" - Median: {data[col].median()}")
    print(f" - Min: {data[col].min()}")
    print(f" - Max: {data[col].max()}")
    print(f" - Range: {data[col].max() - data[col].min():.2f}")
    print()
```

```python
# use describe 
data.describe()
```

### Q-8

Boxplot

```python
plot_data = (data.set_index('species')
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns={0: 'size', 'level_1': 'measurement'})
)

plot_data.head()
```

```python
# using pandas make a boxplot of each petal and sepal measurements

f = plt.figure(figsize=(10, 8))

sns.boxplot(data=plot_data, x='measurement', y='size', hue='species')
```
